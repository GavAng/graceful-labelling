```python
import networkx as nx

from graphs import path, cycle
from graphs.tools import draw_labelled_graph, draw_slideshow
from graphs.graceful_labellings import find_graceful_labellings, find_alpha_valuations
```

```python
edge_sets = [
    cycle(4),
    [(1, 2), (2, 0), (3, 4), (4, 0), (5, 6), (6, 0)],
    cycle(5),
    path(5),
]
for edges in edge_sets:
    print(f"Graph {edges}")
    graph = nx.Graph(edges)
    try:
        graceful_graph = next(find_graceful_labellings(graph))
    except StopIteration:
        print("No graceful labellings found.")
    else:
        draw_labelled_graph(graceful_graph)
```

```python
P_5 = nx.Graph(path(5))
layout = {i: (i, 0) for i in range(5)}
draw_slideshow(find_graceful_labellings(P_5), layout=layout)
```

```python
draw_slideshow(find_alpha_valuations(P_5), layout=layout)
```
