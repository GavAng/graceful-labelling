```python
import networkx as nx

from graphs.tools import draw_labelled_graph
from graphs.graceful_labellings import find_graceful_labellings

edge_sets = [
    # {(0, 1), (1, 2), (2, 3), (3, 0)},
    {(1, 2), (2, 0), (3, 4), (4, 0), (5, 6), (6, 0)},
]
for edges in edge_sets:
    graph = nx.Graph(edges)
    layout = nx.spring_layout(graph)
    for graceful_graph in find_graceful_labellings(graph):
        draw_labelled_graph(graceful_graph, layout=layout)
```
